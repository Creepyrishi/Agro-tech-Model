```python
import pandas as pd
import numpy as np
from  dbfread import DBF
import random
```

```python
dbf = DBF('Nepal forest incidents/data/nepal_activefire.dbf')
df = pd.DataFrame(iter(dbf))
```

```python
df.head()
```

```python
df['ACQ_DATE'] = pd.to_datetime(df['ACQ_DATE'], format='%m/%d/%y')
```

```python
df_after_16 = df[df['ACQ_DATE'] > '2016-01-01']
df.head()
```

```python
df_after_16.info()
```

```python
df_after_16.reset_index()
selected_rows = []
for i in range(1000):
    index = random.randint(0, 1000)
    selected_rows.append(df_after_16.iloc[index])
```

```python
df_subset = pd.DataFrame(selected_rows)
df_subset.head()
```

```python
df_subset.info()
```

```python
df_subset.reset_index()
df_subset.head()
```

```python
not_need_columns = [
    "OBJECTID", 
    "BRIGHTNESS", 
    "SCAN", 
    "TRACK", 
    "SATELLITE", 
    "CONFIDENCE", 
    "VERSION", 
    "BRIGHT_T31", 
    "VDC", 
    "PA", 
    "GRIDID",      
    "LCCODE"
]
df_subset = df_subset.drop(columns=not_need_columns , axis=1)
```

```python
df_subset.to_excel('subset.xlsx')
```
